```python
import pandas as pd
```

```python
filename = "../rankings.csv"
df = pd.read_csv(filename)
```

```python
df = df.drop_duplicates()
df['created_date'] = [x[0:10] for x in df['created_date']]
df['total_supply'] = [int(x) for x in df['total_supply']]
df['market_cap'] = [int(x) for x in df['market_cap']]

# Manually update crypto punks w/ 7 day average
df.loc[1, 'floor_price'] = 75.7363
```

```python
cols = [
    'collection_slug', 
    'created_date', 
    'total_supply', 
    'num_owners', 
    'floor_price', 
    'market_cap', 
    'thirty_day_volume'
]

df_slim = df[cols]
```

### Sorted by market cap

```python
df_market_cap = (
    df_slim
        .query('num_owners > 100')
        .sort_values('market_cap', ascending=False)
        .head(20)
)

df_market_cap.to_csv('data/top_collections_market_cap.csv')

df_market_cap
```

### Sorted by floor price

```python
df_floor_price = (
    df_slim
        .query('num_owners > 100')
        .sort_values('floor_price', ascending=False)
        .head(20)
)

df_floor_price.to_csv('data/top_collections_floor_price.csv')

df_floor_price
```

### Sorted by 30 day volume

```python
df_30D_volume = (
    df_slim
        .query('market_cap > 200')
        .sort_values('thirty_day_volume', ascending=False)
        .head(20)
)

df_30D_volume.to_csv('data/top_collections_30D_volume.csv')

df_30D_volume
```
